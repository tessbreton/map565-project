```python
import pandas as pd

df = pd.read_csv('../data/consumption.csv', delimiter=';')
df.columns = ['date', 'consumption'] # Rename the columns
df['date'] = pd.to_datetime(df['date'], format="%d/%m/%Y")

df.head(3)
```

```python
import matplotlib.pyplot as plt

# Tracer le graphique
plt.figure(figsize=(10, 6),dpi=100)
plt.plot(df['date'], df['consumption'], linestyle='-', color='b', linewidth=0.5)

# Ajouter des tiquettes et un titre
plt.title('Energy consumption over Time')
plt.xlabel('Date')
plt.ylabel('Consumption')
plt.grid(True)

# Afficher le graphique
plt.show()
```
